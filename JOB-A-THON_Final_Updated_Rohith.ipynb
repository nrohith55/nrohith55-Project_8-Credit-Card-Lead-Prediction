```python
from sklearn.preprocessing import scale
# Scaling is done in order to keep all the values in same range so that we can increase the accuracy of the model
```

```python
X=scale(x)
```

```python
X
```

```python
#To chech the count of value
y.Is_Lead.value_counts() # We can observe target records are not equally distributed i.e data is imbalanced.
#So, We have to Balance the target records.For that we are aplying SMOTE method (Synthetic Minority Over-sampling Technique)
```

```python
import seaborn as sns

sns.countplot('Is_Lead',data=y) # Countplot
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=444)
X_train_res,y_train_res=sm.fit_resample(X,y)
```

```python
X_train_res.shape
```

```python
y_train_res.shape
```

```python
X.shape
```

```python
y.shape
```

```python
## Hyper parameter tuning and Model Building
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Creating the hyperparameter grid
c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space}
```

```python
# Instantiating logistic regression classifier
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
  
# Instantiating the GridSearchCV object
logreg_cv = GridSearchCV(logreg, param_grid, cv = 5)
```

```python
logreg_cv.fit(X_train_res,y_train_res.values.ravel())
```

```python
# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_)) 
print("Best score is {}".format(logreg_cv.best_score_))
```

```python
logreg_cv
```

```python
best_random_grid=logreg_cv.best_estimator_
```

```python
best_random_grid
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
y_pred=best_random_grid.predict(X)
print(confusion_matrix(y.values.ravel(),y_pred))
print("Accuracy Score {}".format(accuracy_score(y.values.ravel(),y_pred)))
print("Classification report: {}".format(classification_report(y.values.ravel(),y_pred)))
print('roc_auc_score :',roc_auc_score(y.values.ravel(),y_pred))
```

```python
from scipy.stats import randint
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Creating the hyperparameter grid 
param_dist = {"max_depth": [3, None],
              "max_features": randint(1, 9),
              "min_samples_leaf": randint(1, 9),
              "criterion": ["gini", "entropy"]}
```

```python
# Instantiating Decision Tree classifier
tree = DecisionTreeClassifier()
  # Instantiating RandomizedSearchCV object
tree_cv = RandomizedSearchCV(tree, param_dist, cv = 5)
tree_cv.fit(X_train_res,y_train_res.values.ravel())
```

```python
print("Tuned Decision Tree Parameters: {}".format(tree_cv.best_params_))
print("Best score is {}".format(tree_cv.best_score_))
```

```python
best_random_grid1=tree_cv.best_estimator_
```

```python
best_random_grid1
```
